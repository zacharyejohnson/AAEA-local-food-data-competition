```python
import numpy as np
import pandas as pd 
import seaborn as sn 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("full_column_data.csv", index_col=0)
```

```python
data.dropna(inplace=True)
```

```python
# data must be standardized to mean = 0, sd = 1 
variables = data.keys()
data = StandardScaler().fit_transform(data[data.keys()])
data = pd.DataFrame(data, columns = variables).fillna(value=0)
data
```

```python
pd.DataFrame(np.cov(data))
```

```python
pca = PCA(n_components=10)

prin_comp = pca.fit_transform(data)

pcdf = pd.DataFrame(data = prin_comp)
```

```python
for comp in pcdf: 
    max_corr = 0
    chosen_var = " "
    for var in data: 
        if data[var].corr(pcdf[comp]) > max_corr: 
            max_corr = data[var].corr(pcdf[comp])
            chosen_var = var 
    print("principal component " + str(comp) +" is most correlated with " + chosen_var+ ". \n corrcoef: " + str(max_corr) + "\n\n")
    
    
```
